```python
import numpy as np
from scipy import stats


n_30, ret1_30 = 44699, 20034
n_40, ret1_40 = 45489, 20119


def two_proportion_ztest(x1, n1, x2, n2, label):
    p1, p2 = x1/n1, x2/n2
    p_pool = (x1 + x2) / (n1 + n2)
    se = np.sqrt(p_pool * (1 - p_pool) * (1/n1 + 1/n2))
    z = (p1 - p2) / se
    p_value = 2 * (1 - stats.norm.cdf(abs(z)))
    print(f"{label}: gate_30={p1:.4f}  gate_40={p2:.4f}  diff={p1-p2:.4f}  z={z:.3f}  p={p_value:.5f}")
    return p_value

two_proportion_ztest(ret1_30, n_30, ret1_40, n_40, "Retention_1")
```

```python
ret7_30, ret7_40 = 8501, 8279
two_proportion_ztest(ret7_30, n_30, ret7_40, n_40, "Retention_7")
```

```python
# Casual
two_proportion_ztest(320, 17673, 328, 18316, "Casual")

# Regular
two_proportion_ztest(1911, 15891, 1676, 15497, "Regular")

# Hardcore
two_proportion_ztest(6270, 11135, 6275, 11676, "Hardcore")
```
